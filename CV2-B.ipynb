```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('messi.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

res1 = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)

#OR

height, width = img.shape[:2]
res2 = cv2.resize(img,(2*width, 2*height),interpolation = cv2.INTER_CUBIC)



pfig, axes = plt.subplots(1,3, figsize=(15,10))
axes[0].imshow(img)
axes[1].imshow(res1)
axes[2].imshow(res2)
```

```python
import cv2
import numpy as np

img = cv2.imread('messi.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))

pfig, axes = plt.subplots(1,2, figsize=(15,10))
axes[0].imshow(img,cmap='gray')
axes[1].imshow(dst,cmap='gray')
```

```python
import cv2
import numpy as np

img = cv2.imread('messi.jpg',0)
rows,cols = img.shape


M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))

pfig, axes = plt.subplots(1,2, figsize=(15,10))
axes[0].imshow(img,cmap='gray')
axes[1].imshow(dst,cmap='gray')
```

```python
img = cv2.imread('cross.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('cross.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[20,20],[220,20],[20,220],[220,220]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
import cv2
import numpy as np
img = cv2.imread('sample.jpg',0)
img=256-img
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img,kernel,iterations = 1)
pfig, axes = plt.subplots(1,3, figsize=(15,10))
axes[0].imshow(img,cmap='gray')
axes[1].imshow(erosion,cmap='gray')
axes[2].imshow(dilation,cmap='gray')
```

```python
img = cv2.imread('einput.png',0)
img=256-img
kernel = np.ones((10,10),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(erosion,kernel,iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
pfig, axes = plt.subplots(1,3, figsize=(15,10))
axes[0].imshow(img,cmap='gray')
axes[1].imshow(dilation,cmap='gray')
axes[2].imshow(opening,cmap='gray')
```

```python
img = cv2.imread('dinput.png',0)
img=256-img
kernel = np.ones((10,10),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
erosion = cv2.erode(dilation,kernel,iterations = 1)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
pfig, axes = plt.subplots(1,3, figsize=(15,10))
axes[0].imshow(img,cmap='gray')
axes[1].imshow(erosion,cmap='gray')
axes[2].imshow(closing,cmap='gray')
```

```python

img = cv2.imread('bound.png',0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 2)
output1 = dilation - img
erode = cv2.erode(img,kernel,iterations = 2)
output2 = img - erode
pfig, axes = plt.subplots(1,3, figsize=(15,10))
axes[0].imshow(img,cmap='gray')
axes[1].imshow(output1,cmap='gray')
axes[2].imshow(output2,cmap='gray')
```

```python
img = cv2.imread('hats.png',0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
erosion = cv2.erode(img,kernel,iterations = 1)
grad = dilation - erosion
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
pfig, axes = plt.subplots(1,3, figsize=(15,10))

axes[0].imshow(img,cmap='gray')
axes[1].imshow(grad,cmap='gray')
axes[2].imshow(gradient,cmap='gray')
```

```python
kernel = np.ones((10,10),np.uint8)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
pfig, axes = plt.subplots(1,3, figsize=(15,10))
axes[0].imshow(img,cmap='gray')
axes[1].imshow(tophat,cmap='gray')
axes[2].imshow(blackhat,cmap='gray')
```
