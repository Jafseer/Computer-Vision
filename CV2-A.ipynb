```python
import cv2
import numpy as np

img = cv2.imread('messi.jpg')

# accessing the pixel at (100,100)
px = img[100,100]
print(px)
#[157 166 200]

# accessing only blue pixel
blue = img[100,100,0]
print(blue)
#157


# accessing RED value
px = img.item(10,10,2)
print(px)
#59

# modifying RED value
img.itemset((10,10,2),100)
px = img.item(10,10,2)
print(px)
#100
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#read image
src = cv2.imread('messi.jpg', cv2.IMREAD_UNCHANGED)

src = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)


# extract red channel
red_channel = src[:,:,0]

# create empty image with same shape as that of src image
red_img = np.zeros(src.shape)

#assign the red channel of src to empty image
red_img[:,:,0] = red_channel

fig, axes = plt.subplots(1,3, figsize=(15,10))
axes[0].imshow(src)
axes[1].imshow(red_channel, cmap='gray')
axes[2].imshow(red_img)
```

```python
img = cv2.imread('messi.jpg')

plt.imshow(img)

print(img.shape)
#(342, 548, 3)

print(img.size)
#562248

print(img.dtype)
#uint8

ball = img[280:340, 330:390]

img[273:333, 100:160] = ball

b,g,r = cv2.split(img)
img = cv2.merge((r,g,b))

plt.imshow(img)

fig, axes = plt.subplots(1,3, figsize=(15,10))
axes[0].imshow(b, cmap='gray')
axes[1].imshow(g, cmap='gray')
axes[2].imshow(r, cmap='gray')

```

```python
img1 = cv2.imread('cross.jpg')

img2 = cv2.imread('opencv.png')

dst1 = cv2.addWeighted(img1,0.7,img2,0.3,0)

dst2 = cv2.addWeighted(img1,0.3,img2,0.7,0)

fig, axes = plt.subplots(1,4, figsize=(15,10))
axes[0].imshow(img1)
axes[1].imshow(img2)
axes[2].imshow(dst1)
axes[3].imshow(dst2)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('gradient.png',0)
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('coins.png',0)

ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
            cv2.THRESH_BINARY,11,2)

titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]

for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('flower.jpg',0)

# global thresholding
ret1,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)

# Otsu's thresholding
ret2,th2 = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)


titles = ['Original Image', 'Global Thresholding',
            'OTSU Binarization']
images = [img, th1, th2]

for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
